```python
import pandas as pd
import csv
import os

folder_path = './data1'

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        output_path = os.path.join(folder_path, 'output_1_' + filename)

        df = pd.read_csv(file_path, sep=' ')
        print(f"处理文件：{filename}")
        print(df.head())  # 查看数据是否正确分列

        temp_output_path = os.path.join(folder_path, 'temp_' + filename)
        df.to_csv(temp_output_path, index=False, header=False)
        with open(temp_output_path, 'r', encoding='utf-8') as file:
            reader = csv.reader(file)
            split_rows = []

            for row in reader:
                split_data = row[0].split(',')
                split_rows.append(split_data)

        with open(output_path, 'w', encoding='utf-8', newline='') as file:
            writer = csv.writer(file)
            for row in split_rows:
                writer.writerow(row)

        print(f"处理完成，数据已保存到新文件：{output_path}")

        os.remove(temp_output_path)
```

```python
import pandas as pd
import numpy as np
import os

folder_path = './output_1_data'
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        output_path = os.path.join(folder_path, 'handwash_' + filename)

        df = pd.read_csv(file_path)

        num_rows = len(df)
        results = []

        for i in range(0, num_rows, 100):
            chunk = df.iloc[i:i+100]
            if len(chunk) < 100:
                break  # 如果不足100行，跳过
            mean_values = chunk.mean()
            std_values = chunk.std()
            results.append([mean_values[1], std_values[1], mean_values[2], std_values[2], mean_values[3], std_values[3]])

        results_df = pd.DataFrame(results, columns=['mean_x', 'std_x', 'mean_y', 'std_y', 'mean_z', 'std_z'])
        results_df['Activity'] = 'hand_wash'
        results_df.to_csv(output_path, index=False)

        print(f"处理完成，数据已保存到新文件：{output_path}")
```

```python
#assignment2
import pandas as pd
import numpy as np
import os

input_folder_path = './output_1_data'
output_folder_path = './feature_data_slice1'

# Ensure the output folder exists
os.makedirs(output_folder_path, exist_ok=True)

for filename in os.listdir(input_folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(input_folder_path, filename)
        output_path = os.path.join(output_folder_path, 'handwash_slice1' + filename)

        df = pd.read_csv(file_path)

        num_rows = len(df)
        results = []

        for i in range(0, num_rows, 100):
            chunk = df.iloc[i:i+100]
            if len(chunk) < 100:
                break  # Skip if the chunk has fewer than 100 rows

            # Calculate mean, std, median, and RMS for x, y, z columns
            mean_values = chunk.mean()
            std_values = chunk.std()
            median_values = chunk.median()
            rms_values = np.sqrt((chunk**2).mean())

            results.append([
                mean_values[1], std_values[1], median_values[1], rms_values[1],
                mean_values[2], std_values[2], median_values[2], rms_values[2],
                mean_values[3], std_values[3], median_values[3], rms_values[3]
            ])

        # Create DataFrame with the new columns for mean, std, median, and RMS for x, y, z
        results_df = pd.DataFrame(results, columns=[
            'mean_x', 'std_x', 'median_x', 'rms_x',
            'mean_y', 'std_y', 'median_y', 'rms_y',
            'mean_z', 'std_z', 'median_z', 'rms_z'
        ])
        results_df['Activity'] = 'hand_wash'
        results_df.to_csv(output_path, index=False)

        print(f"处理完成，数据已保存到新文件：{output_path}")
```

```python
import pandas as pd
import os

folder_path = './featured_data_slice1'

dataframes = []

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)

output_path = os.path.join(folder_path, 'combined.csv')
combined_df.to_csv(output_path, index=False)

print(f"所有文件已合并到：{output_path}")
```

```python
#assignment2
import pandas as pd
import os

# Define folder paths and sheet names
folders = {
    'feature_data_slice1': 'slice1',
    'feature_data_slice2': 'slice2',
    'feature_data_slice3': 'slice3',
    'feature_data_slice4': 'slice4'
}

# Create an Excel writer to save each combined DataFrame to a different sheet
output_excel_path = './combined_slices.xlsx'
with pd.ExcelWriter(output_excel_path) as writer:
    for folder_name, sheet_name in folders.items():
        folder_path = f'./{folder_name}'
        dataframes = []
        
        # Read and combine CSV files within each folder
        for filename in os.listdir(folder_path):
            if filename.endswith('.csv'):
                file_path = os.path.join(folder_path, filename)
                df = pd.read_csv(file_path)
                dataframes.append(df)
        
        # Concatenate all DataFrames in the folder and save to the designated sheet
        combined_df = pd.concat(dataframes, ignore_index=True)
        combined_df.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"所有文件夹的数据已合并到单个Excel文件的四个工作表中：{output_excel_path}")
```
